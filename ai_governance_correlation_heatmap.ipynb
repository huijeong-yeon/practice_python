# Correlation Matrix of AI Governance Variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Data construction

```python
data = pd.DataFrame({
    'Country': ['EU', 'US', 'China', 'Korea', 'UK', 'Japan'],
    'Centralization': [3, 1, 5, 3, 2, 2],         
    'Transparency': [5, 2, 1, 4, 3, 3],           
    'Accountability': [5, 2, 2, 3, 2, 2],         
    'Innovation_Support': [3, 5, 4, 4, 3, 4],     
    'Risk_Management': [5, 3, 3, 4, 3, 3],        
    'Stakeholder_Engagement': [5, 2, 1, 3, 3, 3], 
    'Technological_Capacity': [3, 4, 5, 5, 3, 4]  
})

df = pd.DataFrame(data)
```

## 2. Select numeric columns only

```python
numeric_df = df.drop(columns=['Country'])
```

## 3. Compute the Pearson correlation matrix

```python
corr_matrix = numeric_df.corr()
```

## 4. Plot the correlation heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,        
    fmt=".2f",         
    cmap="vlag",      
    cbar_kws={"label": "Correlation"}
)
plt.title("Correlation Matrix of AI Governance Variables")
plt.tight_layout()
plt.show()
```
